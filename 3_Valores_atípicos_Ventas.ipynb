Vamos a visualizar el boxplot de todas las columnas cuantitativas del dataset a la vez para no tener que realizar 

```python
#Realizamos diagrama de caja o bigote
cuant = df.drop(["Unnamed: 0","indice_tiempo"], axis = 1)

fig = plt.figure(figsize =(10, 6))
plt.boxplot(cuant,labels = cuant.columns)
plt.xticks(rotation = 90)
plt.show() #dibujamos el diagrama
```

### Ventas Precio corrientes



```python
#Realizamos histograma de variables cuantitativas
columnas = list(cuant.columns)

indices = [1,2,4,8,9,12,13,15,16,19,21,22,23]
for i in range(len(indices)):
  indices[i] = indices[i]-1
indices

fig = plt.figure(figsize =(8, 3))
plt.hist(x=cuant[columnas[indices[0]]], color='blue', rwidth=0.50)
plt.title('Histograma de Ventas Precio corrientes')
plt.ylabel('Frecuencia')

plt.show() #dibujamos el histograma
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuant[columnas[indices[0]]]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
print()
outliers_iqr= cuant[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
display(outliers_iqr)
print()
data_clean_iqr1= cuant[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
display(data_clean_iqr1)
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr1[columnas[indices[0]]]) 
plt.show() #dibujamos el diagrama
```
